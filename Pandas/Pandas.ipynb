```python
import json
import datetime 
from pandas import Timestamp
```

```python
def convert(obj):
    if isinstance(obj, (datetime.date, datetime.datetime)):
        return obj.isoformat()
    raise TypeError(f"Object of type {obj.__class__.__name__} is not JSON serializable")
```

```python
daily_total_sales.index = daily_total_sales.index.astype(str)
```

```python
sales_summary = {
    "store_sales": store_sales.to_dict(),
    "high_value_transactions": high_value.to_dict(orient="records"),
    "average_quantity_per_product": means.to_dict(),
    "product_sales": product_sales.to_dict(),
    "daily_total_sales": daily_total_sales.to_dict(),
    "total_by_store": total_by_store.to_dict(),
    "total_by_product": total_by_product.to_dict()
}
file_name = "sales_summary.json"
try:
    with open(file_name, "w") as json_file:
        json.dump(sales_summary, json_file, indent=4, default=convert)
    print(f"File successfully created: {file_name}")
except Exception as e:
    print(f"An error occurred while writing the file: {e}")
```
