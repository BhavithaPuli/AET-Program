```sql
SELECT
  (SELECT COUNT(*) FROM vnr3240.default.taxi_raw_records) AS bronze_count,
  (SELECT COUNT(*) FROM vnr3240.default.flagged_rides) AS flagged_count,
  (SELECT COUNT(*) FROM vnr3240.default.weekly_stats) AS weekly_rows;
```

```sql
--Shows which columns have missing data issues
SELECT
  SUM(CASE WHEN id IS NULL THEN 1 END) AS null_id,
  SUM(CASE WHEN pickup_datetime IS NULL THEN 1 END) AS null_pickup,
  SUM(CASE WHEN fare_amount IS NULL THEN 1 END) AS null_fare,
  SUM(CASE WHEN trip_distance IS NULL THEN 1 END) AS null_distance,
  SUM(CASE WHEN passenger_count IS NULL THEN 1 END) AS null_passengers
FROM vnr3240.default.taxi_raw_records;
```

```sql
--checks if trip distance is too long
SELECT *
FROM vnr3240.default.taxi_raw_records
WHERE trip_distance > 9
ORDER BY trip_distance DESC;
```

```sql
--checks all rides when fare is higher than 30
SELECT *
FROM vnr3240.default.taxi_raw_records
WHERE fare_amount > 30
ORDER BY fare_amount DESC;
```

```sql
--validating tables which exist in UI
SHOW TABLE EXTENDED IN vnr3240.default LIKE 'flagged_rides';
SHOW TABLE EXTENDED IN vnr3240.default LIKE 'weekly_stats';
SHOW TABLE EXTENDED IN vnr3240.default LIKE 'top_3';
```

```sql
--Coordinates must fall within NYC boundaries.
SELECT *
FROM vnr3240.default.taxi_raw_records
WHERE pickup_latitude NOT BETWEEN 40 AND 42
   OR pickup_longitude NOT BETWEEN -75 AND -72;
```

```sql
--Check if all Gold records have matching Silver weekly stats.
SELECT *
FROM vnr3240.default.top_3
WHERE avg_fare IS NULL OR avg_distance IS NULL;
```

```sql
--total output
SELECT
  (SELECT COUNT(*) FROM vnr3240.default.taxi_raw_records) AS raw,
  (SELECT COUNT(*) FROM vnr3240.default.flagged_rides) AS flagged,
  (SELECT COUNT(*) FROM vnr3240.default.weekly_stats) AS weeks,
  (SELECT COUNT(*) FROM vnr3240.default.top_3) AS gold_output;
```
